# Jumlah Coffee Berdasarkan Tahun

```python
print(df.columns)
```

```python
tahun = df.groupby('Year').count()[['Coffee']].sort_values(by='Coffee',ascending=True).reset_index()
tahun = tahun.rename(columns={'Coffee':'numberOfCoffee'})
```

```python
fig = plt.figure(figsize=(15,5))
sns.barplot(x=tahun['Year'], y=tahun['numberOfCoffee'], color='royalblue')
plt.xticks(rotation=60)
```

# seleksi fitur

```python
features = ['Year','Cocoa','Tea']
x = df[features]
y = df['Coffee']
x.shape, y.shape
```

# split data training dan data testing

```python
from sklearn.model_selection import train_test_split
x_train, X_test, y_train, y_test = train_test_split(x,y,random_state=70)
y_test.shape
```

# membuat model regresi linier

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train,y_train)
pred = lr.predict(X_test)
```

```python
score = lr.score(X_test, y_test)
print('akurasi model regresi linier = ', score)
```

# membuat inputan model regresi linier

```python

input_data = np.array([[2024,500,700]])

prediction = lr.predict(input_data)
print('Estimasi harga komoditas kategori minuman Coffee ($/Kg) :', prediction)
```

# save model

```python
import pickle

filename = 'estimasi_Harkom_Coffee.sav'
pickle.dump(lr,open(filename,'wb'))
```
